```python
def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:

        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.replace(r"[\u200b\t\r\n]", "", regex=True)
            .str.replace(" +", " ", regex=True)
            .str.title()
        )
    return df
```

```python
def reporte_calidad_global(dfs, nombres):
    """
    Crea un resumen de calidad de varios DataFrames.

    Parámetros:
      dfs: lista de DataFrames (por ejemplo [ventas_clean, clientes_clean, marketing_clean])
      nombres: lista de nombres correspondientes (["VENTAS", "CLIENTES", "MARKETING"])
    """
    resumen = []

    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated(keep=False).sum()
        columnas = len(df.columns)
        filas = len(df)

        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)

    return reporte
```

```python
ventas_perf = (
    ventas_clean
    .assign(
        ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]

    )
)
```

```python
resumen_prod = (
    ventas_perf

    .groupby(
        by="producto",
        dropna=False,
        as_index=False,
        observed=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
    )
)
```

```python
p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,
    interpolation="linear"
)


ventas_top = (
    resumen_prod
    .query(
        "ingreso_total >= @p80_ingreso",
        engine="python"
    .sort_values(
        by=["ingreso_total", "unidades"],
        ascending=[False, False],
        na_position="last",
        ignore_index=False

    )
)

print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("✅ Productos de ALTO RENDIMIENTO (top 20% por ingreso):")
display(ventas_top.head(60))
```

# 📘 Presentación Final — Conclusiones del Proyecto

El presente proyecto tuvo como objetivo aplicar de manera integral las distintas etapas del análisis de datos sobre un caso práctico vinculado a **ventas y marketing**, culminando en una presentación final orientada a la comunicación de hallazgos e insights ante un panel simulado de expertos de **SynthData**.

A lo largo del trabajo, los datos analizados permitieron identificar que los **ingresos no se distribuyen de forma homogénea entre los productos**, sino que existe una concentración significativa en un grupo reducido de artículos de alto rendimiento. Este comportamiento quedó evidenciado tanto en las medidas estadísticas descriptivas como en las visualizaciones realizadas, donde se observó una **asimetría positiva** y la presencia de **outliers relevantes**, interpretados como productos exitosos y no como errores de carga.

El análisis exploratorio mostró que la mayoría de los productos se agrupa en un rango de ingresos medio, mientras que algunos pocos superan ampliamente ese nivel, generando un impacto significativo en el ingreso total. Este hallazgo constituye un insight clave para la toma de decisiones, ya que pone de manifiesto la importancia de identificar y potenciar los productos líderes dentro de una estrategia comercial.

Asimismo, el estudio de la relación entre el **precio promedio y las unidades vendidas** evidenció una **correlación positiva débil**, lo que indica que el precio no actúa como un factor limitante directo de las ventas. Esto sugiere que variables como el valor percibido, la funcionalidad y la relevancia del producto para el consumidor influyen más en la demanda que el precio en sí mismo, aportando información valiosa para el diseño de estrategias de marketing y posicionamiento.
